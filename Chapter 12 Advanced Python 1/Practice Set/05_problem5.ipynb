Store the multiplication tables generated in problem 3 in a file named Tables.txt.

```python
take_value = int(input("Enter any number: "))
table = [take_value * i for i in range(1,10+1)]

with open(f"tables/{take_value}.txt","a")as f:
    f.write(f"Table of {take_value} is {str(table)}")
    f.close()
```
