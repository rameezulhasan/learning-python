**Map**

```python
numbers = [1,2,3,4,5]

squared = map(lambda x: x**2, numbers)
print(list(squared))
```

```python
names = ["Rameez", "Ali", "Hassan"]

upper_names = map(str.upper,names)
print(list(upper_names))
```

**Filter**

```python
numbers = [1,2,3,4,5,6,7]
odd_numbers = filter(lambda data:data%2 != 0, numbers)
print(list(odd_numbers))
```

```python
names = ["ram", "rameez", "ali", "hassan"]
check_length = filter(lambda x:len(x) >= 4, names)
print(list(check_length))
```

**Reduce**

```python
from functools import reduce
numbers = [1,2,3,4,5]
sum_total = reduce(lambda x,y: x+y, numbers)
print(sum_total)
```

```python
multiply_total = reduce(lambda x,y: x*y, numbers)
print(multiply_total)
```
