Write a program to input name, marks and phone number of a student and format it
using the format function like below:

“The name of the student is Harry, his marks are 72 and phone number is 99999888”

```python
take_name = input("Enter Your Name: ")
take_marks = int(input("Enter Your Marks: "))
take_phoneNo = input("Enter Your Phone Number: ")

print("The name of the student is {0}, his marks are {1} and phone number is {2}".format(take_name,take_marks,take_phoneNo))
```
